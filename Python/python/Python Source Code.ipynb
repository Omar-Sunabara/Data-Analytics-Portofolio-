```python
# Question 1 - What are the regional sales in the best performing country? (Region)

import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
file_path = r"C:\Users\c\Desktop\Question1 - Region.csv"
data = pd.read_csv(file_path)

# Adjust the bar chart for clarity with actual sales figures on the y-axis
plt.figure(figsize=(12, 6))

# Plot bars for each region
bars = plt.bar(data['Region'], data['RegionSales'], color='blue', edgecolor='black')

# Add title and axis labels
plt.title('Regional Sales', fontsize=16)
plt.xlabel('Region', fontsize=12)
plt.ylabel('Sales (in $)', fontsize=12)

# Rotate x-axis labels for better readability
plt.xticks(rotation=60, fontsize=10)

# Format y-axis with actual sales figures (no scientific notation)
plt.ticklabel_format(style='plain', axis='y')

# Add value labels on top of each bar
for bar in bars:
    plt.text(
        bar.get_x() + bar.get_width() / 2,  # X position (center of the bar)
        bar.get_height(),                   # Y position (top of the bar)
        f"${bar.get_height():,.0f}",       # Format the value
        ha='center', va='bottom',          # Align text
        fontsize=10, color='black'         # Font size and color
    )

# Tight layout for clear spacing
plt.tight_layout()

# Display the bar chart
plt.show()
```

```python
# Question 2: What is the relationship between annual leave taken and bonus?

# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Pandas used to read csv file
data = pd.read_csv(r"C:\Users\Safar\OneDrive\Desktop\Q2.csv")
print(data.head())

# Correlation coefficient calculated between annual leave and bonus
correlation = data[['VacationHours', 'Bonus']].corr()
correlation_coef = correlation.iloc[0, 1]
print(correlation_coef)
print('The correlation coefficient between annual leave taken and bonus is -0.04')

# Scatter plot of the data
plt.scatter(data['VacationHours'], data['Bonus'], label='Data Points')

# Used numpy's polyfit function to calculate the line of best fit (1 for linear)
slope, intercept = np.polyfit(data['VacationHours'], data['Bonus'], 1)

# Created the line of best fit
plt.plot(data['VacationHours'], slope * data['VacationHours'] + intercept, color='red', label='Line of Best Fit')

plt.title('Annual leave vs. Bonus')
plt.xlabel('Vacation Hours')
plt.ylabel('Bonus in $')

plt.legend()
plt.show()
```

```python
# Question 3 - What is the relationship between Country and Revenue? (Countries)

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Reading a CSV file
df = pd.read_csv(r"C:\Users\c\Desktop\Question3 - Countries.csv")
print(df.head())
print(df.info())

# Selecting country name and revenue and sorted by descending order
df1 = df[['Country', 'Revenue']].sort_values(by='Revenue', ascending= False)  
print(df1)

# Creating a barchart with country name and revenue 
plt.bar(df1.Country, df1.Revenue, color='red')
plt.title('Revenue by Country', fontsize=18, fontweight='bold', color='navy')

plt.xlabel('Country', fontsize=12)
plt.ylabel('Revenue', fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--')
plt.ticklabel_format(style='plain', axis='y')
plt.tight_layout()
plt.show()
```

```python
# Question 5 - What is the relationship between store trading duration and revenue?

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the CSV file
file_path = r"C:\Users\c\Desktop\Question5.csv"
data = pd.read_csv(file_path)

# Convert to DataFrame
df = pd.DataFrame(data)

# Calculate the correlation coefficient
correlation_coefficient = np.corrcoef(data['TradingDurationYears'], data['Revenue'])[0, 1]
print(f"The correlation coefficient is: {correlation_coefficient}")

# Scatter plot for Trading Duration vs Revenue
plt.figure(figsize=(10, 6))
plt.scatter(df['TradingDurationYears'], df['Revenue'], alpha=0.75)

# Add labels and title
plt.title('Relationship Between Store Trading Duration and Revenue', fontsize=14)
plt.xlabel('Trading Duration (Years)', fontsize=12)
plt.ylabel('Revenue ($)', fontsize=12)

# Show grid
plt.grid(True, alpha=0.5)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# Question 6 - What is the relationship between the size of the stores, number of employees and revenue?

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

# Reading the CSV file with pandas
df = pd.read_csv(r"C:\Users\c\Desktop\Q6.csv")
print(df.head())
print(df.info())

# selecting Square feet and Revenue
df1 = df[['Squarefeet', 'Revenue']]
print(df1)

# Correlation of Square feet and Revenue
correlation1 = df1['Squarefeet'].corr(df1['Revenue'])
print(f"Correlation coefficient: {correlation1}")

# Scatterplot of Square feet and Revenue
plt.figure(figsize=(10, 6))
plt.scatter(df1.Squarefeet, df1.Revenue, color='red')

# Line of best fit
m, b = np.polyfit(df1.Squarefeet, df1.Revenue, 1) 
plt.plot(df1.Squarefeet, m * df1.Squarefeet + b, color='blue', label='Best fit line')

plt.title('Revenue by Store Size', fontsize=18, fontweight='bold', color='navy')
plt.xlabel('Square Feet', fontsize=16)
plt.ylabel('Annual Revenue', fontsize=16)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()

# Selecting Revenue by Employee Number
df2 = df[['Employees', 'Revenue']]
print(df2)

# Correlation of Revenue by Employee Number
correlation2 = df2['Employees'].corr(df1['Revenue'])
print(f"Correlation coefficient: {correlation2}")

# Scatterplot of Revenue by Employee Number
plt.figure(figsize=(10 , 6))
plt.scatter(df2.Employees, df2.Revenue, color='red', alpha=0.5, label='Data points')

# Line of best fit
m, b = np.polyfit(df2.Employees, df2.Revenue, 1)  # m is slope, b is y-intercept
plt.plot(df2.Employees, m * df2.Employees + b, color='blue', label='Best fit line')

plt.title('Revenue by Employee Number', fontsize=18, fontweight='bold', color='navy')
plt.xlabel('Employees', fontsize=16)
plt.ylabel('Annual Revenue', fontsize=16)
plt.grid(axis='y', linestyle='--')
plt.tight_layout()
plt.show()


# Correlation of all three variables
df1 = df[['Squarefeet', 'Revenue']]  # Squarefeet and Revenue
df2 = df[['Employees', 'Revenue']]  # Employees and Revenue
df3 = df[['Squarefeet', 'Employees', 'Revenue']]  # Squarefeet, Employees, and Revenue

correlation_matrix = df3.corr()
print(correlation_matrix)


# Heatmap of the 3 variables with correlation coefficients
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', cbar=True)
plt.title('Correlation of Squarefeet, Employees, Revenue', fontsize=16)
plt.show()

# Scatterplot of all 3 variables (Squarefeet,Revenue and Employees)
fig = px.scatter(df, x='Squarefeet', y='Revenue', size='Employees', color='Employees',
                 title='Square Feet vs Annual Revenue',
                 labels={'Squarefeet': 'Square Feet', 'Revenue': 'Annual Revenue'},
                 hover_data=['Employees'])
fig.show()
```
